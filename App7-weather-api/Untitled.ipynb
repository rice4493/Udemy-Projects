# calculate a new column out of an existing column 

```python
import numpy as np
df["TG0"] = df['   TG'].mask(df['   TG'] == -9999, np.nan)
df
```

```python
df["TG"] = df["TG0"]/10
```

```python
df
```

```python
df["Fahrenheit"] = df["TG"]*(9/5) + 32
```
